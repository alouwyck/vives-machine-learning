## **Opgave**

1. Download [hier](https://archive.ics.uci.edu/ml/datasets/Zoo) de **ZOO dataset**, lees die in en verken de data. Er zijn 101 samples en 17 attributen. Attribuut 'animal name' heb je niet nodig. Er zijn geen ontbrekende waarden. Er zijn wel twee 'frog' examples, maar de ene is 'venomous' (giftig) en de andere niet. Dus je moet beide examples behouden! Je hoeft de dataset *niet* op te splitsen. M.a.w. je gebruikt telkens de volledige dataset om een model te trainen (en houdt zo geen rekening met mogelijke overfitting!).

2. Pas **KNN** toe en probeer een zo hoog mogelijke nauwkeurigheid te halen bij het voorspellen van feature *type*. Omdat we de dataset niet opsplitsen, moet *K groter zijn dan 2*! Pas op! De dataset is niet gebalanceerd, dus we gebruiken de [**balanced accuracy**](https://scikit-learn.org/stable/modules/model_evaluation.html#balanced-accuracy-score) (met default input parameters). Maak ook een plot van de bijhorende *confusion matrix*.

3. Pas **KMEANS** toe om de dataset in clusters op te delen. Probeer ook hier een zo hoog mogelijke *'balanced accuracy'* te halen wanneer je de clusters vergelijkt met de klassen volgens feature *type*. Maak ook hier een plot van de confusion matrix. Pas op! Je hoeft niet per se de dataset op te splitsen in 7 clusters! Je zal ook een methode moeten zoeken om de cluster labels te hernummeren zodat je ze kan vergelijken met de chte labels uit de dataset!

4. Bereken het rekenkundig gemiddelde van de twee 'balanced accuracies' die je hebt behaalt bij KNN en KMEANS. Print die mooi uit. **Winnaar is diegene die de hoogste gemiddelde score haalt!**

5. Presenteer je resultaat: leg uit hoe je te werk bent gegaan en toon de confusion matrix van je finale KNN en KMEANS model. En uiteraard ook de bijhorende 'balanced accuracy' en je eindscore. Plot en print alles mooi uit!
