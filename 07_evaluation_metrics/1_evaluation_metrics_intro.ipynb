## Groepsfuncties

Met groepsfuncties bedoelen we functies die de elementen in een bepaalde dimensie gaan groeperen of aggregeren. Bijvoorbeeld de som nemen van alle elementen in elke kolom van een 2D matrix.

In NumPy heb je vaak ook methodes met dezelfde naam die juist hetzelfde doen. Nemen we `sum` als voorbeeld en passen de functie en de methode toe op een vector met 4 enen:

````python
v = np.ones(4)
v
````

````python
np.sum(v)
````

````python
v.sum()
````

Nemen we nu een 3 x 4 matrix, dit is een 2D matrix met 3 rijen en 4 kolommen:

````python
m = np.ones((3, 4))
m
````

````python
np.sum(m)
````

````python
m.sum()
````

We kunnen echter ook de as specificeren waarlangs de functie of methode moet opereren. We doen dit door de parameter `axis` mee te geven:

````python
np.sum(m, axis=0)
````

````python
m.sum(axis=0)
````

Dus `axis=0` zorgt ervoor dat de som van de elementen van elke kolom wordt genomen.

image.png

En `axis=1`? Inderdaad, dit telt de getallen van elke rij op:

````python
np.sum(m, axis=1)
````

````python
m.sum(axis=1)
````

image.png

Merk op dat de aggregatie ervoor zorgt dat het aantal dimensies met 1 verminderd wordt. We kunnen echter parameters `keepdims` op `True` zetten, en dan behouden we 2 dimensies:

````python
m.sum(axis=0, keepdims=True)
````

````python
np.sum(m, axis=1, keepdims=True)
````

## Arrays concateneren

Arrays met overeenkomstige dimensies kan je samenvoegen met de functie `concatenate`:

````python
A = np.ones((3, 4))
B = np.zeros((3, 2))
np.concatenate((A, B), axis=1)
````

De twee matrices die je wil samenvoegen stop je dus in een tuple en dan kan je de axis meegeven. Is `axis=1` dan voeg je de matrices horizontaal samen, dus langs de as van de kolomindices. Is `axis=0`, dan voeg je de matrices verticaal samen:

````python
C = np.zeros((2, 4))
np.concatenate((A, C), axis=0)
````

## Confusion matrix

Wanneer we een confusion matrix hebben dan is het wel handig om de sommen van de rijen en kolommen te berekenen.

````python
c = np.array([[45, 5], [2, 48]])
c
````

````python
c = np.concatenate((c, c.sum(axis=1, keepdims=True)), axis=1)
c
````

````python
c = np.concatenate((c, c.sum(axis=0, keepdims=True)), axis=0)
c
````

De relatieve aantallen:

````python
c = c / c[-1, -1]
c
````

Ook handig om weten is de functie `diag()` waarmee we de elementen op de diagonaal kunnen selecteren:

````python
c = np.array([[13, 2], [1, 14]])
c
````

````python
np.diag(c)
````

De accuracy kunnen we dus eenvoudig als volgt berekenen:

````python
np.diag(c).sum() / c.sum() * 100
````
