## Wat zijn recommender systems?

**Recommender systems** zijn software tools of technieken die gebruikers voorzien van zinvolle aanbevelingen. 'Aanbevelingen' in het Engels zijn 'recommendations' en dus noemt men zo'n tools ook wel 'recommendation systems'. In plaats van 'systems' spreekt men ook soms van 'platforms' of 'engines'.

Typische voorbeelden zijn de systemen die Amazon, Facebook, Google, LinkedIn, Netflix, YouTube, Spotify, Bol.com en Booking.com toepassen om gebruikers suggesties te doen over wat men het best kan kopen, welke muziek of films men ook leuk zal vinden, welke artikels men interessant zal vinden om te lezen, welke reizen men het best kan boeken, of welke personen het best bij iemand passen.

De figuur hieronder geeft op een eenvoudige manier weer hoe een recommender system werkt:
- Eerst en vooral zijn er de **gebruikers**, die een profiel hebben met attributen zoals woonplaats, leeftijd, gender, enz.
- **Items** is de algemene naam voor wat men aanbeveelt, bv. songs in het geval van spotify, films in het geval van Netflix, nieuwe vrienden in het geval van Facebook, en reisaccommodaties in het geval van Booking.com. Ook die items kunnen allerhande attributen hebben.
- Dan zijn er ook nog de **transacties**, de interacties tussen de gebruikers en het recommendation system. We denken dan in eerste plaats aan **ratings** waarmee een gebruiker aangeeft hoe leuk die een bepaald item vindt. Maar ook het klikken op links van producten of het plaatsen van bestellingen zijn voorbeelden van transacties. Het zijn dus in feite gelogde data die het systeem ook opslaat in een databank.
- Op basis van al deze data, nl. gebruikersprofielen, items en transacties, gaat het recommender system aanbevelingen of **recommendations** doen. Die kunnen al dan niet gepersonaliseerd zijn.
- Sommige recommender systems kunnen ook de **context** in rekening brengen. In feite zijn dit attributen van de transacties. We denken aan het tijdstip, de dag in de week, het weer, de locatie, de gemoedstoestand van de gebruiker, enzovoort. Een dergelijk 'context-aware' systeem zal - hopelijk - een ander restaurant aanbevelen voor een business lunch met een klant dan voor een romantisch diner.

image.png

Als we nu eens focussen op de oranje rechthoek in de figuur hierboven, dan kunnen we verschillende **algoritmes** onderscheiden die op de inputdata worden toegepast. Net als bij andere machine learning taken is de output van elk van die algoritmen een **model** waarmee **voorspellingen** worden gedaan. Als de transacties bijvoorbeeld bestaan uit ratings die gebruikers hebben gegeven aan bepaalde film, dan zal dat model de ratings voorspellen van nieuwe films die gebruikers nog niet hebben bekeken. Op basis van die voorspelde ratings worden dan **aanbevelingen** gedaan, bijv. de 5 meest interessante films voor een bepaalde gebruiker. Als het systeem **gepersonaliseerd** is dan krijgt elke gebruiker een eigen top 5 te zien. In een **niet-gepersonaliseerd** systeem krijgt elke gebruiker dezelfde top 5 te zien. Ook *association rules* leveren niet-gepersonaliseerde aanbevelingen op. Dat is dus het eerste belangrijke onderscheid dat we moeten maken, zoals op de figuur hieronder is te zien.

image.png

Wat de gepersonaliseerde systemen betreft zijn de twee belangrijkste types algoritmen content-based filtering en collaborative filtering. Er zijn uiteraard nog andere soorten algoritmes, zoals hybride systemen die beiden toepassen, maar deze vallen buiten de scoop van dit vak.

Bij **content-based filtering** worden items aanbevolen die grote overeenkomst vertonen met items waarin een gebruiker in het verleden geïnteresseerd was. Het vergelijken van items doet men op basis van item attributen.

Bij **collaborative filtering** zijn er geen attributen van items nodig en kijkt men enkel naar de ratings die gebruikers aan items geven. Die **ratings** kunnen **expliciet** zijn, zoals scores die men geeft aan bepaalde producten, of **impliciet**, zoals het klikken op een product. Sommige collaborative filtering algoritmes kijken naar de gelijkenissen tussen items, die noemen we **item-based**, terwijl de **user-based** algoritmen naar de gelijkenissen tussen verschillende gebruikers kijken. **Matrix Factorization** algoritmen passen meer geavanceerde wiskundige technieken toe, zoals singuliere-waardenontbinding. En dan zijn er nog andere systemen die zelfs gebruik maken van deep learning of reinforcement learning.   

## Belangrijke matrices

In deze sectie bespreken we 2 belangrijke matrices:

- de **item-content matrix** (ICM) die de hoofdrol speelt in *content-based* filtering
- en de **user-rating matrix** (URM) of **user-item matrix** die in *collaborative* filtering heel vaak  wordt gebruikt.

Beide matrices bevatten vaak heel wat nullen. Matrices waarvan het grootste deel van hun elementen nul zijn, noemen we ijl of **sparse**.

Een derde belangrijke matrix is de **similarity matrix** die de gelijkenissen bevat tussen items of users. Die komt aan bod in het stuk over collaborative filtering.

### Item-conent matrix

De item-content matrix is eenvoudigweg een tabel met items en bijhorende attributen. De items vormen de rijen, de attributen de kolommen.

image.png

Neem als voorbeeld een dataset met films. De items zijn de films en attributen kunnen bijvoorbeeld acteurs en actrices zijn die al dan niet meespelen in een film. Op die manier krijgen we een matrix met nullen en enen. Maar in plaats van een binaire matrix, kan het ook een matrix zijn met gewichten. Tom Cruise is de hoofdrolspeler in Top Gun en krijgt daarom een gewicht van 100%, maar Stan Lee speelde een bijrol in The Avengers en krijgt om die reden bijvoorbeeld maar een gewicht van 50% voor die film.

image.png

### User-rating matrix

De user-rating matrix bevat de ratings die gebruikers aan items hebben gegeven. De rijen van deze matrix komen overeen met de users, de kolommen met de items. De elementen van de matrix zijn dan de **ratings**. Die kunnen **expliciet** zijn, bijv. het aantal sterren dat een film heeft gekregen van gebruikers, of **impliciet**, bijv. of een gebruiker een film al dan niet heeft bekeken.

Dit wordt hieronder visueel voorgesteld. De user-rating matrix heeft symbool $R$ gekregen. Rij $u$ komt overeen met user $u$ en kolom $i$ met item $i$. Het element op rij $u$ en in kolom $i$ is rating $r_{ui}$. Dit is de rating die user $u$ aan item $i$ heeft gegeven. In het voorbeeld van een expliciete rating zal $r_{ui}$ het aantal sterren zijn die gebruiker $u$ heeft gegeven aan film $i$, en dus krijgen we de waarden 1, 2, 3, 4 of 5. In het voorbeeld van een impliciete rating zal $r_{ui}$ gelijk zijn aan 1 als gebruiker $u$ film $i$ heeft gezien. Als impliciete rating kan men ook de tijd nemen dat een gebruiker naar een film heeft gekeken. Andere voorbeelden van impliciete ratings zijn het aantal keren dat iemand een song heeft beluisterd, of het aantal producten dat een klant heeft besteld.

image.png

### Sparse matrices

User-rating matrix $R$ zal uiteraard ook veel nullen bevatten, want gebruikers zullen nog heel veel films niet beoordeeld hebben of niet bekeken. Een matrix met heel veel nullen noemen we een ijle matrix, in het Engels een **sparse matrix**. Een matrix met weinig of geen nullen noemt men 'dense'. Het percentage elementen die niet nul zijn, de dichtheid of 'density', is bij user-rating matrices meestal kleiner dan 0.01%. Die van Netflix bijvoorbeeld is slechts ongeveer 0.002%! Soms gebruikt men ook de term 'sparsity' wat het percentage nullen is in de matrix. Het spreekt voor zich dat 'density' en 'sparsity' samen 100% zijn. Met $n$ het totaal aantal elementen in de matrix en $n_z$ het aantal elementen gelijk aan nul:

> sparsity = $\frac{n_z}{n}$

> density = $\frac{n - n_z}{n}$

> density + sparsity = $\frac{n - n_z}{n} + \frac{n_z}{n} = \frac{n - n_z + n_z}{n} = \frac{n}{n} = 1$

Nuttig om weten is dat **SciPy** een **module `sparse`** heeft om 'sparse' matrices te creëren en te bewerken: https://docs.scipy.org/doc/scipy/reference/sparse.html. Sparse matrices gaan enkel de elementen opslaan die niet nul zijn. Er zijn verschillende manieren om dat te doen. Als voorbeeld nemen wij het compressed sparse row format:

````python
import numpy as np
from scipy import sparse

A = np.zeros((10, 10))  # 10 x 10 matrix met nullen
A[2, 3] = 1  # element (2, 3) gelijkstellen aan 1
A[5, 7] = np.pi  # element (5, 7) gelijkstellen aan pi
A = sparse.csr_matrix(A)  # dense numpy array omzetten naar sparse csr matrix
A
````

Wanneer we `A` effectief printen, dan worden enkel de elementen verschillend van nul weergegeven samen met hun index:



````python
print(A)
````

## Non-personalized recommendations

Niet gepersonaliseerde aanbevelingen gelden voor alle gebruikers. M.a.w. elke user krijgt dezelfde aanbevelingen. Afhankelijk van de data waarover we beschikken, kunnen we verschillende technieken toepassen.

### Nearest neighbors

We kunnen het 'nearest neighbors' algoritme toepassen om de $K$ items te vinden die het meest "gelijken" op een gegeven item $i$ wanneer een item-content matrix is gegeven. Nagaan in hoeverre twee items op elkaar gelijken, doen we door het berekenen van de afstand tussen die twee items.

Nemen we als eenvoudig voorbeeld de item-content matrix in file "movies_actors.csv". We lezen de file in en passen `NearestNeighbors` toe uit Scikit-Learn. Let op! We nemen de **unsupervised** versie van het algoritme omdat we in dit geval geen labels hebben!

````python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
````

We lezen de file in:

````python
movies = pd.read_csv("movies_actors.csv", sep=";")
movies
````

We hebben enkel de kolommen met acteurs en actrices nodig:

````python
X = movies.drop(labels="Title", axis=1).values
X
````

Stel dat we de film willen vinden die meest lijkt op de laatste film "Unbreakable", dan moeten we de volledige dataset fitten met uitzondering van die laatste film:

````python
NN = NearestNeighbors(n_neighbors=1)  # n_neighbors = 1 omdat we maar 1 film willen
NN.fit(X[:-1, :]);
````

Die laatste film geven we daarna door aan methode `kneigbors`:

````python
distance, index = NN.kneighbors(X[[-1], :])  # [-1] omdat de submatrix ook 2D moet zijn!
````

Het eerste outputargument `distance` is de afstand, die standaard de Euclidische afstand is. Met het tweede outputargument `index` kunnen we dan de film selecteren die het meest lijkt op "Unbreakable":

````python
movies.Title[index[0][0]]
````

We krijgen als resultaat "Pulp Fiction" omdat net als in "Unbreakable" ook "Samuel L. Jackson" en "Bruce Willis" een rol spelen in die film.

### Most popular and best rated

Als we over een user-rating matrix beschikken, dan kunnen we de meest populaire items en de 'best rated' items selecteren en aan de gebruikers aanbevelen.

We nemen een eenvoudig voorbeeld, nl. de dataset uit file "movies_ratings.csv". Die bevat expliciete ratings van 12 gebruikers voor 14 films. De rating variëren van 1 t.e.m. 5:

````python
import pandas as pd

ratings = pd.read_csv("movies_ratings.csv", sep=";")
ratings
````

We halen de user-rating matrix uit deze dataset door enkel de kolommen met titels van films over te houden en de `NaN` waarden te vervangen door `0`. We zetten het dataframe ook om naar een NumPy array met integer waarden:  

````python
R = ratings.drop(ratings.columns[0], axis=1).fillna(0).values.astype(int)
R
````

Om de meest populaire items te krijgen gaan we voor elke film na hoeveel gebruikers een rating hebben gegeven. We moeten dus voor elk item het aantal ratings bepalen dat niet nul is:

````python
N = (R > 0).sum(axis=0)  # aantal ratings per item dat niet nul is
i = N.argmax()  # index van het item met het meeste ratings niet gelijk aan nul
ratings.columns[i+1]  # i+1 want eerste kolom bevat de kijkers!
````

Film "Bad Boys" heeft inderdaad het meeste ratings, nl. 8. Maar er zitten ook enkele minder positieve ratings tussen van 1 en 2. Daarom kunnen we beter de gemiddelde rating per film berekenen. Pas op! We mogen enkel het gemiddelde berekenen van de ratings die niet nul zijn!

> $\bar{r}_{i} = \frac{\sum_{u} r_{ui}}{N_i}$

met $\bar{r}_{i}$ de gemiddelde rating voor item $i$ en $N_i$ het aantal ratings $r_{ui}$ voor item $i$ die niet nul zijn.

Laten we de gemiddelde rating voor alle films eens berekenen en dan de film selecteren die de hoogste gemiddelde rating heeft:

````python
best_rated = R.sum(axis=0) / N  # gemiddelde rating van elk item
i = best_rated.argmax()  # index van het item met de hoogste gemiddelde rating
ratings.columns[i+1]  # i+1 want eerste kolom bevat de kijkers!
````

We zien dat "Titanic" de hoogste gemiddelde rating heeft. Maar "Titanic" heeft maar 1 rating van 5 gekregen, dus echt betrouwbaar is die uitkomst niet.

Om het effect van het aantal ratings ook te laten doorwegen maken we gebruik van een **shrink term**. Dat is een constante $C$ die we optellen bij het aantal ratings $N_i$:

> $\bar{r}_{i} = \frac{\sum_{u} r_{ui}}{N_i + C}$

In dit geval is $\bar{r}_{i}$ de **shrinked average rating** voor item $i$.

Laten we dat eens toepassen op onze dataset voor $C=1$:

````python
C = 1
best_rated_C1 = R.sum(axis=0) / (N + C)
i = best_rated_C1.argmax()
ratings.columns[i+1]  # i+1 want eerste kolom bevat de kijkers!
````

Nu krijgen we "Men in Black III". En wat als we $C=2$ nemen?

````python
C = 2
best_rated_C2 = R.sum(axis=0) / (N + C)
i = best_rated_C2.argmax()
ratings.columns[i+1]  # i+1 want eerste kolom bevat de kijkers!
````

Dan krijgen we "Pulp Fiction". En dat blijft zo als we de shrink term nog groter nemen. Dus hier zouden we "Pulp Fiction" als de "best rated" film kunnen aanbevelen.

## Collaborative filtering in the neighborhood

Gepersonaliseerde aanbevelingen zijn op maat van individuele gebruikers. De twee belangrijkste groepen van algoritmes zijn **content-based filtering** en **collaborative filtering**. In beide gevallen proberen we om de ontbrekende ratings van gebruikers in te schatten of te voorspellen. Op basis van die geschatte ratings kunnen we dan voor elke gebruiker persoonlijke aanbevelingen doen.

In dit hoofdstuk bespreken we een vorm van collaborative filtering die we **neighborhood-based filtering** noemen, en zich baseert op de gelijkenissen tussen gebruikers of items. Zoals we al hebben gezien spreken we van **user-based** of **item-based**.

Omdat er in praktijk meestal heel veel items en users zijn, gaat men meestal enkel gebruik maken van de items of users die het meest gelijken op het item en de user van de rating die men wil voorspellen. Men gaat dus een *nearest-neighbor* techniek toepassen om het probleem te vereenvoudigen.

````python
import numpy as np
import pandas as pd
````

````python
ratings = pd.read_excel("movie_ratings.xlsx", index_col=0)  # we nemen de eerste kolom als index (= rijnaam)
ratings
````

Om praktische redenen stoppen we de gebruikers in een list `users` en de films in een list `items`. De matrix zelf zetten we om naar een NumPy array `R` met floats.

````python
users = list(ratings.index)
items = list(ratings.columns)
R = ratings.values.astype('float')
````

### Stap 1: de similarity matrix opstellen

Als we een item-based voorspelling willen doen van de ontbrekende ratings, dan moeten we een similarity matrix `S` opstellen die de gelijkenis tussen de films aangeeft. Zoals we reeds hebben gezien is de **cosinusgelijkenis** interessant in dit geval. Toepassen van deze 'similarity measure' geeft ons een **vierkante, symmetrische matrix**. Op de diagonaal krijgen we elementen die allemaal één zijn omdat een item 100% lijkt op zichzelf.

Laten we eerst eens de cosinusgelijkenis berekenen tussen de eerste 2 films: "Men In Black" en "Star Trek":

````python
i1 = items.index("Men In Black")  # index van Men In Black
i2 = items.index("Star Trek")  # index van Star Trek

r1 = R[:, i1]  # ratings van Men In Black
r2 = R[:, i2]  # ratings van Star Trek

np.dot(r1, r2) / np.sqrt(np.sum(r1**2) * np.sum(r2**2))
````

Dat was niet zo moeilijk. Maar hoe moeten we te werk gaan als we de cosinusgelijkenis tussen "Men In Black" en "Brave Heart" willen berekenen? De kolom van die laatste film bevat nl. 2 `NaN`-waarden en de uitkomst zal dus ook `NaN` zijn. We mogen dus enkel de rijen nemen die geen `NaN` bevatten. Dat kunnen we testen met NumPy functie `isnan`:

````python
i1 = items.index("Men In Black")  # index van Men In Black
i2 = items.index("Brave Heart")  # index van Brave Heart

r1 = R[:, i1]  # ratings van Men In Black
r2 = R[:, i2]  # ratings van Brave Heart

b = ~(np.isnan(r1) | np.isnan(r2))
r1, r2 = r1[b], r2[b]  # ratings van MIB en BH die voor beiden niet NaN zijn!

np.dot(r1, r2) / np.sqrt(np.sum(r1**2) * np.sum(r2**2))
````

We kunnen de cosinusgelijkenis normaliseren of centreren door de gemiddelde rating van de items af te trekken van elke rating:

````python
i1 = items.index("Men In Black")  # index van Men In Black
i2 = items.index("Star Trek")  # index van Star Trek

r1 = R[:, i1].copy()  # ratings van Men In Black - copy nemen, want kolom wordt aangepast!
r2 = R[:, i2].copy()  # ratings van Star Trek - copy nemen, want kolom wordt aangepast!

r1 -= r1.mean()  # ratings voor MIB centreren = gemiddelde rating voor MIB aftrekken
r2 -= r2.mean()  # ratings voor BH centreren = gemiddelde rating voor BH aftrekken

np.dot(r1, r2) / np.sqrt(np.sum(r1**2) * np.sum(r2**2))
````

Wanneer we de ratings voor beide films centreren, dan krijgen we een negatieve waarde, wat erop wijst dat de films niet op elkaar lijken, en zelfs eerder tegenovergestelde ratings hebben: kijkers die "Men In Black" goed vinden, zullen "Star Trek" niet zo goed vinden, en omgekeerd:

````python
print(np.round(np.hstack((r1[:, np.newaxis], r2[:, np.newaxis])), 2))
````

Wanneer we nu "Men In Black" gaan vergelijken met "Brave Heart", dan hebben we 2 opties om de gemiddelde rating voor "Men In Black" te berekenen:

1. ofwel nemen we alle ratings van deze film
2. ofwel nemen we enkel de ratings van users die ook "Brave Heart" hebben beoordeeld.

In het laatste geval gaan we dus enkel het gemiddelde nemen van de rijen die geen `NaN`-waarden bevatten. Laten we de twee opties eens vergelijken:

````python
# optie 1: we nemen alle ratings van Men in Black:

i1 = items.index("Men In Black")  # index van Men In Black
i2 = items.index("Brave Heart")  # index van Brave Heart

r1 = R[:, i1].copy()  # ratings van Men In Black
r2 = R[:, i2].copy()  # ratings van Brave Heart

r1 -= r1.mean()  # gemiddelde aftrekken voor het verwijderen van de NaN-rijen
r2 -= np.nanmean(r2)  # nanmean gebruiken, anders is de uitkomst NaN!

b = ~(np.isnan(r1) | np.isnan(r2))
r1, r2 = r1[b], r2[b]  # ratings van MIB en BH die voor beiden niet NaN zijn!

np.dot(r1, r2) / np.sqrt(np.sum(r1**2) * np.sum(r2**2))
````

````python
# optie 2: we nemen enkel de ratings uit de rijen die geen NaN bevatten:

i1 = items.index("Men In Black")  # index van Men In Black
i2 = items.index("Brave Heart")  # index van Brave Heart

r1 = R[:, i1].copy()  # ratings van Men In Black
r2 = R[:, i2].copy()  # ratings van Brave Heart

b = ~(np.isnan(r1) | np.isnan(r2))
r1, r2 = r1[b], r2[b]  # ratings van MIB en BH die voor beiden niet NaN zijn!

r1 -= r1.mean()  # gemiddelde aftrekken Na het verwijderen van de NaN-rijen
r2 -= r2.mean()  # nanmean is niet nodig hier want de NaNs zijn verwijderd

np.dot(r1, r2) / np.sqrt(np.sum(r1**2) * np.sum(r2**2))
````

Het is duidelijk dat de twee opties tot andere resultaten leiden. Sommige auteurs noemen optie 1 de "centered" of "adjusted cosine similarity" en optie 2 de "Pearson correlation coefficient". Dat is dus het subtiele verschil tussen beiden.

We definiëren nu een functie om de cosinusgelijkenissen tussen alle items te berekenen, en voorzien ook de optie om te normaliseren en zo de Pearson correlatiecoëfficiënt als resultaat te krijgen.

````python
def cosine_similarity(R, normalize=False):
    # R is de user-item matrix met de users in de rijen en de items in de kolommen
    # als normalize=True, dan wordt de Pearson correlatiecoëfficiënt berekend
    # default is normalize=False
    nusers, nitems = R.shape  # aantal users en aantal items
    rated = ~np.isnan(R)  # boolean matrix die aangeeft welke users welke items een rating hebben gegeven
    S = np.nan * np.ones((nitems, nitems))  # initialiseer similarity matrix S met NaNs
    S[np.diag_indices_from(S)] = 1.0  # de waarden op de diagonaal moeten 1 zijn
    for i1 in range(nitems):  # doorloop alle items
        for i2 in range(i1 + 1, nitems):  # maak gebruik van de symmetrie!
            b = rated[:, i1] & rated[:, i2]  # de user moet beide items i1 en i2 beoordeeld hebben!
            if np.any(b):  # er moeten ratings zijn om de gelijkenis te kunnen berekenen
                r1 = R[b, i1]  # ratings van i1 (= kolom i1)
                r2 = R[b, i2]  # ratings van i2 (= kolom i2)
                if normalize:  # normaliseren = gemiddelde item rating aftrekken van de ratings
                    r1 -= np.mean(r1)
                    r2 -= np.mean(r2)
                S[i1, i2] = S[i2, i1] = np.dot(r1, r2) / np.sqrt(np.sum(r1**2) * np.sum(r2**2))  # cosinusgelijkenis
    return S
````

Laten we nu de cosine similarity matrix `S` berekenen met deze functie:

````python
S = cosine_similarity(R)
print(np.round(S, 3))
````

De Pearson correlatiematrix is:

````python
Sp = cosine_similarity(R, normalize=True)
print(np.round(Sp, 3))
````

Om de centered cosine similarity matrix te krijgen, moeten we gewoon de gecentreerde rating matrix doorgeven aan functie `cosine_similarity`:

````python
Rc = R - np.nanmean(R, axis=0, keepdims=True)  # gecentreerde rating matrix = R - gemiddelde item ratings
Sc = cosine_similarity(Rc)
print(np.round(Sc, 3))
````

Als we de Pearson correlatiematrix vergelijken met de gecentreerde cosinusgelijkenismatrix, dan zien we dat enkel de derde en vierde rij en kolom andere waarden hebben, omdat de derde en vierde kolom in de rating matrix `NaN`-waarden bevatten.

Eén van deze matrices kan nu gebruikt worden om item-based collaborative filtering toe te passen. Willen we user-based werken, dan moeten we de gelijkenissen tussen users berekenen. Dat kan op dezelfde manier gebeuren als items vergelijken, alleen moeten we nu rijen i.p.v. kolommen nemen. Laten we eens zien hoe gelijkend "Pietro" en "Ekaterina" zijn wat betreft hun beoordelingen van de films in de dataset:

````python
u1 = users.index("Pietro")  # index van Pietro
u2 = users.index("Ekaterina")  # index van Ekaterina

r1 = R[u1, :]  # ratings van Pietro = rij u1
r2 = R[u2, :]  # ratings van Ekaterina = rij u2

np.dot(r1, r2) / np.sqrt(np.sum(r1**2) * np.sum(r2**2))
````

De berekening gebeurt inderdaad op dezelfde manier, alleen nemen we rijen `u1` en `u2` ipv kolommen `i1` en `i2`. Dus eigenlijk kunnen we ook onze functie `cosine_similarity` gebruiken om user-user similarity matrices te berekenen, alleen moeten we de transpose van de rating matrix doorgeven:

````python
print(np.round(cosine_similarity(R.T), 3))  # R.T doorgeven!
````

### Stap 2: de neighborhood bepalen

Om nu te voorspellen welke rating een gebruiker zal geven aan een film, bekijkt men gewoonlijk de zogenaamde neighborhood van de film:

1. men neemt enkel de films waarvan de gelijkenis met de gegeven film groter is dan een opgegeven drempel;

en/of

2. men neemt enkel de K films waarvan de gelijkenis met de gegeven film het grootst is.

Nemen we als voorbeeld de film "Brave Heart" omdat die 2 ontbrekende ratings heeft. Laten we de Pearson correlatiematrix `Sp` nemen die we daarnet hebben opgesteld:

````python
i = items.index("Brave Heart")  # kolom index van Brave Heart
sp = Sp[:, i]
print(np.round(sp[:, np.newaxis], 3))  # kolom selecteren uit correlatiematrix
````

Passen we een drempel toe, bijv. de correlatie moet groter zijn dan 0.5, dan houden we 2 films over (omdat we de correlatie met zichzelf niet meenemen natuurlijk):

````python
threshold = 0.5  # drempel
j = np.arange(len(sp))  # indices van de items/films
j = np.delete(j, np.where(j == i))  # index van BH deleten
j = j[sp[j] >= threshold]  # indices houden van films met correlatie > 0.5
j
````

We krijgen inderdaad de indices van de laatste 2 films. Willen we bijvoorbeeld de 3 meest gelijkende films overhouden, dan moeten we de correlaties sorteren en de 3 hoogste waarden nemen:

````python
K = 3  # 3 films die meest gelijken op Brave Heart
j = np.arange(len(sp))  # indices van de items/films
j = np.delete(j, np.where(j == i))  # index van BH deleten
j = j[np.argsort(sp[j])]  # indices sorteren volgens correlatiecoëfficiënt (van klein naar groot)
j = j[:-K-1:-1]  # K laatste indices overhouden (want grootste waarden zitten achteraan de array)
j
````

Omdat we de 3 meest gelijkende films nemen, komt hier nu ook nog de tweede film erbij.

### Stap 3: de ontbrekende ratings voorspellen

Eens we de ratings hebben van de films die het meest lijken op de film die nog niet werd beoordeeld door een gebruiker, kunnen we makkelijk een voorspelling maken van de rating die de gebruiker aan die film zal geven. We kunnen hier 2 methoden toepassen:

- **classificatie**: we nemen de rating over die het meest voorkomt in de neighborhood;
- **regressie**: we nemen het gewogen gemiddelde van de ratings in de neighborhood, met de similarities als gewichten.

image.png

In het geval van regressie moeten we de volgende berekening maken om te schatten welke rating $\hat{r}$ user $u$ zal geven aan item/film $i$:

> $\hat{r}_{ui} = \frac{\sum_{j\in N(i)} S_{ij} r_{uj}}{\sum_{j\in N(i)} S_{ij}}$

Die formule ziet er vrij heftig uit, maar zegt eigenlijk gewoon dat de voorspelde rating $\hat{r}$ de gewogen is van alle ratings $r_{uj}$ die de user $u$ heeft gegeven aan de films $j$ uit de neighborhood $N$ van film $i$. De gewichten zijn de similarities $S_{ij}$ tussen de films $j$ en film $i$.

Willen we voorspellen welke rating "Jesper" aan "Brave Heart" zal geven, dan kunnen we gebruik maken van de K films die we daarnet hebben bepaald:

````python
u = users.index("Jesper")  # index van user "Jesper"
r_uj = R[u, j]  # de ratings die Jesper gaf aan de 3 films j die meest op Brave Heart lijken
print("ratings 3 dichtste films:", r_uj)
S_ij = Sp[i, j]  # de correlaties tussen de 3 films j en Brave Heart i
rp_ui = np.dot(S_ij, r_uj) / np.sum(S_ij)
print("voorspelde rating:", rp_ui)
````

Als we de Pearson correlatiematrix of de centered cosine similarity matrix gebruiken, dan kunnen we beter ook de ratings normaliseren die we gebruiken om de voorspellingen te berekenen:

> $\hat{r}_{ui} = \mu_i + \frac{\sum_{j\in N(i)} S_{ij} r_{uj}-\mu_j}{\sum_{j\in N(i)} S_{ij}}$

met $\mu_i$ het gemiddelde van alle ratings voor item $i$ en $\mu_j$ het gemiddelde van alle ratings voor item $j$.

De ratings centreren we dus door er het gemiddelde van af te trekken. Maar omdat we met genormaliseerde ratings rekenen, voorspellen we dus ook een genormaliseerde rating, en moeten we bij die voorspelling ook de gemiddelde rating van item $i$ optellen:


````python
u = users.index("Jesper")  # index van user "Jesper"
r_uj = R[u, j]  # de ratings die Jesper gaf aan de 3 films j die meest op Brave Heart lijken
print("ratings 3 dichtste films:", r_uj)
mu_j = np.mean(R[:, j], axis=0)  # gemiddelde rating van de 3 films j
print("gemiddelde rating van de 3 dichtste films:", mu_j)
mu_i = np.nanmean(R[:, i])
print("gemiddelde rating van film Brave Heart:", mu_i)
S_ji = Sp[j, i]  # de correlaties tussen de 3 films j en Brave Heart i
rp_ui = mu_i + np.dot(S_ji, r_uj - mu_j) / np.sum(S_ji)
print("voorspelde rating:", rp_ui)
````

Stappen 2 en 3 - de neighborhood bepalen en de voorspelling berekenen - kunnen we nu in één functie `predict` stoppen:

````python
def predict(R, S, u, i, threshold=None, K=None, with_means=False):
    # R is de user-item matrix met de users in de rijen en de items in de kolommen
    # S is de item-item similarity matrix
    # u is de rijindex van de gebruiker voor wie een rating wordt voorspeld
    # i is de kolomindex van het item waarvoor een rating wordt voorspeld
    # threshold is de minimale similarity
    # K is het aantal gelijkaardige items dat wordt meegenomen in de voorspelling
    # with_means geeft aan of de ratings gecenterd moeten worden of niet
    #  default is False (niet centeren)
    j = np.arange(R.shape[1])  # item rating indices
    j = np.delete(j, np.where((j == i) | np.isnan(R[u, j])))  # verwijder item i en items zonder rating
    s = S[:, i]  # item similarities
    if threshold is not None:  # threshold gegeven
        j = j[s[j] >= threshold]  # enkel ratings nemen van items met similarity >= threshold
    if K is not None:  # K gegeven
        j = j[np.argsort(s[j])[:-K-1:-1]]  # neem de K items met de grootste similarity
    if len(j) > 0:  # er moeten uiteraard ratings overblijven om te kunnen voorspellen...
        if with_means:  # R normaliseren
            rmean = np.nanmean(R, axis=0)
            return rmean[i] + np.dot(S[j, i], R[u, j] - rmean[j]) / np.sum(S[j, i])
        else:
            return np.dot(S[j, i], R[u, j]) / np.sum(S[j, i])
    else:
        return None
````

Laten we eens testen door opnieuw de rating voor Brave Heart van Jesper in te schatten:

````python
print("Geschatte rating van Jesper voor Brave Heart:")
print("- zonder normalisatie:", predict(R, Sp, u, i, K=3))
print("- met normalisatie:", predict(R, Sp, u, i, K=3, with_means=True))
````

De functie is opnieuw geschreven vanuit het oogpunt van item-based filtering. Willen we user-based werken, dan moeten we dus input matrix `R` met de ratings transponeren, een user-user similarity matrix `S` opgeven, en we moeten ook `u` en `i` omwisselen! D.w.z. dat we de index van het item aan `u` moeten toekennen de user id aan `i`.

Willen we bijvoorbeeld de rating van Sara voor Ace Ventura inschatten o.b.v. gelijkenissen tussen gebruikers, dan doen we dat als volgt (als we niet normaliseren):

````python
predict(R.T, cosine_similarity(R.T),
        u=items.index("Ace Ventura"), i=users.index("Sara"),
        threshold=0)
````

Om onze eigen dataset om te zetten naar een gepast dataset object van `Surprise`, moeten we eerst het dataframe omzetten naar een nieuw dataframe met kolommen `user`, `item` en `rating`. We schrijven daarvoor een functie:

````python
def ratings_to_surprise_df(ratings):
    df = dict(user=[], item=[], rating=[])
    for _, row in enumerate(ratings.iterrows()):
        user = row[0]
        for item, rating in row[1].items():
            if not np.isnan(rating):  # enkele ratings die niet NaN zijn!
                df['user'].append(user)
                df['item'].append(item)
                df['rating'].append(int(rating))
    return pd.DataFrame(data=df)
````

### Cold start problem

Wanneer we met een nieuwe gebruiker te maken hebben, een gebruiker die nog geen ratings heeft, dan is de rij in rating matrix $R$ die overeenkomt met die user leeg, waardoor we de similarity met andere users niet kunnen bepalen, en dus geen voorspellingen kunnen maken voor die user via collaborative filtering. Naar analogie met een wagen die niet start wanneer het koud is, staat dit probleem bekend als "the cold start problem".

Er zijn allerhande oplossingen mogelijk. Men kan bijvoorbeeld een nieuwe gebruiker vragen een profiel aan te maken of diens profiel van een andere site overnemen. Men kan ook gebruik maken van association rule learning of content-based filtering. En er zijn ook meer geavanceerde methodes mogelijk die bijv. gebruik maken van deep learning.

Bij nieuwe items krijgen we trouwens een gelijkaardig probleem: doordat hun kolom in matrix $R$ leeg is, kunnen we de gelijkenis met andere items in de dataset gaan bepalen.

## Matrix Factorization

Matrix factorization is eveneens een methode binnen de **collaborative filtering** familie. Hierbij gaat men de user rating matrix factoriseren of ontbinden. We spreken in dat geval van **latent factors** die we kunnen interpreteren als verborgen features van users en items. In het geval van films zouden we kunnen denken aan filmgenres die niet gekend zijn, hoewel die factoren niet altijd een 'echte' betekenis hebben.

We bespreken één populaire techniek die gekend staat als **Funk SVD** en gebaseerd is op de singuliere-waardenontbinding uit de lineaire algebra.  

### SVD

SVD staat voor **singular-value decomposition**, singuliere-waardenontbinding in het Nederlands. Het is een techniek uit de lineaire algebra om een matrix te ontbinden. We gaan niet ingaan op de wiskundige details van deze methode, maar demonstreren ze gewoon eens a.d.h.v. de user-item matrix uit file "movie_ratings.xlsx".

````python
import pandas as pd
import numpy as np

ratings = pd.read_excel("movie_ratings.xlsx", index_col=0)
ratings
````

We zetten het dataframe om naar een NumPy array en we vervangen de `NaN`-waarden door nul, omdat de SVD anders niet zal werken.

````python
R = ratings.values
R[np.isnan(R)] = 0.0
R
````

Het is nu mogelijk om matrix $R$ te ontbinden in 3 matrices:

>$ R = U \Sigma V^T$

met:
- $R$ de $n_u$ x $n_i$ rating matrix;
- $U$ een $n_u$ x $n_u$ matrix;
- $V$ een $n_i$ x $n_i$ matrix;
- $\Sigma$ een $n_u$ x $n_i$ diagonaal matrix met de singuliere waarden;
- $n_u$ het aantal users;
- $n_i$ het aantal items.

Merk op dat we $V^T$ schrijven, de transpose van $V$, omdat de kolommen van $V$ overeenkomen met item features, net zoals de kolommen van $U$ overeenkomen met user features. Die features of factoren zijn als het ware verborgen eigenschappen van users en items, en men spreekt dan ook van **latent factors**. "Latent" betekent immers "niet zichtbaar".


image.png

Met de functie `svd` uit module `linalg` van SciPy kunnen we dat eenvoudig uitvoeren:

````python
from scipy.linalg import svd

U, sv, VT = svd(R)
print(U)
print()
print(sv)
print()
print(VT)
````

Vector `sv` bevat enkel de singuliere waarden. Die moeten we dus op de diagonaal van een matrix met nullen plaatsen:

````python
S = np.zeros(R.shape)
n = len(sv)
S[:n, :n] = np.diag(sv)
print(S)
````

Als we die 3 matrices met elkaar gaan vermenigvuldigen, dan krijgen we opnieuw matrix `R`:

````python
np.round(np.dot(np.dot(U, S), VT))
````

Wat is nu het nut van het toepassen van SVD? Wel, net als bij PCA kunnen we ons probleem reduceren door enkel de factoren in rekening te brengen die een serieuze bijdrage leveren aan de ratings. Hoe vinden we die factoren? Door naar de singuliere waarden te kijken. Die zijn gesorteerd van groot naar klein. Als we nu hun cumulatieve percentages berekenen, dan krijgen we:

````python
print(np.round(sv, 4))  # singuliere waarden
print(np.round(np.cumsum(sv / sv.sum()) * 100, 3))  # cumulatieve percentages
````

De eerste waarde heeft een aandeel van bijna 54%, de eerste twee waarden van iets meer dan 72%, enzovoort. Als we nu enkel de factoren willen meenemen in de voorspelling die samen meer 90% bijdragen, dan mogen we de 2 laatste laten vallen:

````python
from numpy.linalg import multi_dot
r = 4  # enkel de eerste 4 features gebruiken
print(multi_dot((U[:, :r], S[:r, :r], VT[:r, :])))
````

We hebben inderdaad enkel de eerste vier features van `U` en `V` in rekening gebracht, en krijgen zo een benadering van de rating matrix `R`. Wanneer die laatste matrix zeer groot is, kan dat het probleem sterk reduceren.

image.png

Dat is allemaal leuk, maar is het niet de bedoeling om de ontbrekende ratings te voorspellen? Juist! En dus is die SVD niet zo nuttig, tenzij dan dat we het probleem kunnen reduceren, zoals we ook bij PCA doen.

Maar om ontbrekende ratings te kunnen inschatten, moeten we een methode hebben die de `NaN`-waarden effectief buiten beschouwing laat. En dat is exact wat Funk SVD doet.

### Funk SVD

Funk SVD is genaamd naar Simon Funk die deze methode bedacht en toepaste in de legendarische [Netflix Prize](https://en.wikipedia.org/wiki/Netflix_Prize). Hij won de prijs niet, maar toch raakte zijn methode bekend, omdat die ook efficiënt is én relatief eenvoudig. Bovendien deelde hij de werking ervan met het grote publiek via zijn [blog](https://sifter.org/simon/journal/20061211.html). We leggen kort uit hoe de methode werkt en maken dan gebruik van de Surprise library om de methode te demonstreren.  

We weten dat we de rating matrix kunnen ontbinden in een matrix met user features en een matrix met item features. De matrix met voorspelde ratings kunnen we dus als volgt schrijven:

> $ \hat{R} = UV^T$

Matrices $U$ en $V^T$ moeten we vermenigvuldigen. Dat betekent dat we voor elk element in $\hat{R}$ het inwendig product moeten nemen van de overeenkomstige rij in $U$ en kolom in $V^T$:

> $\hat{r}_{ui} = U_{u1}V_{i1} + U_{u2}V_{i2} + ... = \sum_{k=1}^{r}U_{uk}V_{ik}$

met:
- $\hat{r}_{ui}$ de geschatte rating van user $u$ voor item $i$, d.i. het element op rij $u$ en kolom $i$ in matrix $\hat{R}$
- $U_{uk}$ het $k$-de element in rij $u$ van $U$
- $V_{ik}$ het $k$-de element in rij $i$ van $V$
- $r$ het aantal latente factoren

image.png

Bij Funk SVD houden we rekening met de ontbrekende ratings en daarom stellen we enkel de vergelijkingen op voor de gekende ratings $r_{ui}$. We krijgen zo een zeer groot stelsel van vergelijkingen met onbekende parameters $U_{uk}$ en $V_{ik}$. Omdat die parameters onbekend zijn, kunnen we ook beter matrix $\Sigma$ met 'gewichten' weglaten, of beter gezegd, 'wegwerken' in matrices $U$ en $V$.

Dus eigenlijk moeten we ervoor zorgen dat het verschil tussen gekende rating $r_{ui}$ en geschatte rating $\hat{r}_{ui}$ zo klein mogelijk is! Net als bij lineaire regressie kunnen we de **sum of squared errors** (SSE) als loss functie $L$ definiëren:

> $L = \sum_{(u,i) \in N}(r_{ui} - \hat{r}_{ui})^2$

met $N$ de verzameling van gekende ratings $r_ui$ en $|N|$ het aantal gekende ratings. De lossfunctie kunnen we ook hier minimaliseren m.b.v. **stochastic gradient descent**.

Als resultaat van deze minimalisatie krijgen we de optimale waarde voor alle elementen van $U$ en $V$. Eens die elementen gekend zijn, kunnen de ontbrekende ratings worden bepaald met dezelfde formule voor het berekenen van voorspelling $\hat{r}_{ui}$.

We zouden het algoritme opnieuw zelf kunnen implementeren, maar de methode is ook beschikbaar in surprise als klasse `SVD`, en dus gaan we dit gebruiken:

````python
from surprise.prediction_algorithms.matrix_factorization import SVD
from surprise.dataset import Dataset
from surprise.reader import Reader
````

Eerst moeten we het Pandas dataframe omzetten naar een Surprise dataset:

````python
dataset = Dataset.load_from_df(ratings_to_surprise_df(ratings), Reader())
dataset = dataset.build_full_trainset()
````

Dan initialiseren we klasse `SVD`. Het aantal latente factoren `n_factors` zetten we op 4. We moeten dat aantal zelf kiezen, dus is dit een *hyperparameter*!

````python
funk = SVD(n_factors=4,  # aantal latente factoren
           reg_all=0)  # geen regularisatie
````

Het model trainen doen we door methode `fit` aan te roepen en er de gecreëerde dataset aan door te geven:

````python
funk.fit(dataset);
````

Voorspellingen maken we met methode `predict`:

````python
funk.predict(uid="Sara", iid="Ace Ventura").est
````

Dit is Funk SVD in zijn meest eenvoudige vorm. We verwijzen naar de Surprise documentatie voor alle mogelijke opties i.v.m. het in rekening brengen van bias en het toepassen van regularisatie. Het is ook mogelijk om de dataset op te splitsen in een training- en testset, en om cross-validatie toe te passen. Uiteraard kunnen ook evaluatiemetrieken worden berekend, zoals de RMSE:

````python
from surprise import accuracy
predictions = funk.test(dataset.build_testset())
accuracy.rmse(predictions);
````
