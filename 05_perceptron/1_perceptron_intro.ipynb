## Opgave

Het bestand "reviews_python.txt" bevat de reviews van 20 studenten nadat ze het vak "Programming in Python" hadden gevolgd. In het bestand staan de antwoorden op de vraag wat ze van de programmeertaal Python vonden. We lezen het bestand in en krijgen een lijst van 20 zinnen, dus een list van 20 strings.

```python
with open("reviews_python.txt", "r") as txt:
    reviews = txt.read().splitlines()
reviews
```

We gaan de zinnen nu labelen: de zinnen die positief zijn, krijgen `True` als label, de zinnen die negatief zijn `False`. We maken daarvoor gebruik van een dataframe:

```python
import pandas as pd
reviews = pd.DataFrame(dict(
    reviews=reviews,
    happy=[True, True, False, True, False, True, True, True, False, True,
           False, True, False, True, True, True, False, False, False, False]
))
reviews
```

We hebben onze trainingset en we gaan nu proberen te voorspellen of een student die een nieuwe review schrijft tevreden is of niet. Dit is een voorbeeld van **supervised learning**, en meer bepaald van **classificatie**, omdat we gelabelde data hebben, en de feature die we willen voorspellen is categorisch. Maar hoe gaan we hier best te werk?
